# Manipulando el dataframe

Se busca determinar qué valores contiene cada columna, para esto se utiliza sobre el data frame 'value_counts()' que además de mostrar los valores de cada columna, muestra la frecuencia con que aparece cada valor.

Antes de generar correlaciones, análisis y conclusiones, es necesario entender los datos que estamos manejando, como se distribuyen, qué hay de cada tipo, cuantos hay, etc. Y la mejor manera para endenterlo es mediante gráficos.

En primer lugar se exploran las características de los empleados. 

```python
# Se declara la creación de una figura de tamaño 30x30
plt.figure(figsize=(30,30))
plt.style.use('seaborn') # Se utiliza como estilo seaborn, que contiene una visualización más elegante
plt.suptitle("Datos de los empleados",fontsize=30, fontweight='bold', y=.92) # El titulo general de la figura se crea con suptitle de matplotlib, junto con esto se definen sus parámetros, tamaño, tipo de fuente y 'y' que define su posición en el eje y


plt.subplot(3, 2, 1); # Esta figura contiene 3 filas, dos columnas, este subplot es el primero de todos estos
plt.hist(df['Age'], bins = 20, edgecolor='black', color='r', alpha=0.8) # Se crea un histograma con las edades con 20 bins (o divisiones entre clases), de color rojo, con sus bordes negros (con 'edgecolor'), alpha define la transparencia
plt.title("Edades", size=24, pad=14, fontweight='bold') # Se define el titulo, con sus parámetros 
plt.xlabel("Rango edades", fontsize=20)
plt.ylabel("Conteo", fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=15) # Y los tamaños de los ticks


plt.subplot(3, 2, 2); # Segundo sublplot, esta vez un histograma con los valores de la distancia entre el hogar y trabajo de los empleados
plt.hist(df['DistanceFromHome'], bins = 20, edgecolor='black', color='b', alpha=0.8)
plt.title("Distancia del hogar", size=24, pad=14, fontweight='bold')
plt.xlabel("Distancia", fontsize=20)
plt.ylabel("Conteo", fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=15)


plt.subplot(3, 2, 3); # El tercer subplot es un gráfico de torta
plt.pie(df['Education'].value_counts().values, # Grafica los valores de el nivel de educacion de cada empleado
        labels=['Licenciatura', 'Magíster', 'Universitaria', 'No Universitaria', 'Doctorado'], # Con sus respectivas etiquetas *** Es importante que el orden de estas coincidan con lo que en este caso arroja value_counts
        autopct=lambda p: '{:.0f}'.format(p * df['Education'].value_counts().values.sum() / 100), # Para los valores de los labels internos se utiliza lambda con una función que retorne un valor absoluto y no un porcentaje,
                                                              # En este caso corresponde al conteo total de la columna dividido entre 100
        explode=[0.05,0.02,0.02,0.02,0.02], # explode separa una o todas las porciones del gráfico a cierta distancia indicada en la lista respectivamente
        colors=sns.color_palette('Paired'), # los colores del gráfico están en este caso definido por una paleta de color de seaborn
        startangle = 0,
        textprops={'fontsize': 18})
plt.title("Nivel de educación ", size=24, pad=14, fontweight='bold')

# El resto de los subplots son gráficos de torta, para el estado civil, departamento y campo de educación.
plt.subplot(3, 2, 4); 
plt.pie(df['MaritalStatus'].value_counts().values, labels=df['MaritalStatus'].value_counts().index ,autopct=lambda p: '{:.0f}'.format(p * df['MaritalStatus'].value_counts().values.sum() / 100), explode=[0.05,0.02,0.02], colors=sns.color_palette('Accent'), startangle = 180,textprops={'fontsize': 18})
plt.title("Estado Civil ", size=24, pad=14, fontweight='bold')


plt.subplot(3, 2, 5);
plt.pie(df['Department'].value_counts().values, labels=df['Department'].value_counts().index ,autopct=lambda p: '{:.0f}'.format(p * df['Department'].value_counts().values.sum() / 100), explode=[0.05,0.02,0.02], colors=sns.color_palette('Set1'), startangle = 90,textprops={'fontsize': 18})
plt.title("Departamento ", size=24, pad=14, fontweight='bold')


plt.subplot(3, 2, 6);
plt.pie(df['EducationField'].value_counts().values, labels=df['EducationField'].value_counts().index ,autopct=lambda p: '{:.0f}'.format(p * df['EducationField'].value_counts().values.sum() / 100), explode=[0.05,0.02,0.02,0.02,0.02,0.02], colors=sns.color_palette('Set2'), startangle = 180,textprops={'fontsize': 18})
plt.title("Tipo de trabajo", size=24, pad=14, fontweight='bold')
```

# ¿Cómo se relacionan los datos entre sí?

Una vez comprendidas las distribuiciones de los datos, es posible entender muchas más cosas en relación a los empleados, la empresa y su conformación en base a quienes trabajan en ella.


Pero es posible ir más allá, relacionando los datos bajo condiciones que permitan escudriñar aún más en la información que vamos poco a poco entendiendo.

Un buen ejemplo podría ser, ¿cómo se distribuyen los sueldos en base a las características de los empleados?

Por último, y solo para descartar errores en la interpretación, se prodece a crear sampleos de los datos, es decir muestras reducidas de estos generadas a partir de métodos randomicos que posibiliten la interpretación de los datos desde distintas perspectivas.

Estos métodos se utilizan cuando la cantidad de datos es muy grande o cuando no se tiene acceso a la cantidad total de datos, se manipulan estos para replicarlos en mayores cantidades y de esta manera poder generar conjuntos de datos representativos de la realidad. 

Pero en este caso, se realizará para comprobar desde "otra muestra" de datos las inferencias que buscamos.

```python
# Es necesario en primer lugar resetear el índice del dataframe pues al indicar como índice otra columna, esta columna no contiene datos secuenciales
# Se emplea para esto, el método reset_index()
df1 = df1.reset_index()
```

```python
# Se genera primero un sampleo de un 60% de los datos con reemplazo, es decir un mismo dato puede repetirse varias veces 
np.random.seed(0) # Se crea una semilla para random de numpy que guarará los valores aleatoreos generados
for _ in range(int((len(df1)*60)/100)): # Se itera sobre el 60& de los datos, es decir 
    samples = np.random.choice(a=int((len(df)*60)/100), size=int((len(df)*60)/100), replace=False) # Cada iteración selecciona una fila al azar y la guarda bajo un índice que se guardará en la variable sample
bootstrappedDF = df1.loc[samples, :] # bootstrappedDF contiene los datos de df1 a los que se les aplicó el sampleo
bootstrappedDF
```

```python
# Se genera la matriz de correlación para el df con el sampleo aplicado
correlation= bootstrappedDF.corr()
mask = np.triu(np.ones_like(correlation, dtype=bool))
f, ax = plt.subplots(figsize=(30, 30))
cmap = sns.diverging_palette(240, 10, n=9)
sns.heatmap(correlation, annot=True, mask=mask, cmap=cmap, vmax=.3, center=0,
             linewidths=.5, cbar_kws={"shrink": .5})
plt.yticks(fontsize=21)
plt.xticks(fontsize=21)
```

```python
# Nuevamente se crea un sampleo al 60% de los datos, bajo otra semilla, pero esta vez sin el reemplazo activado, es decir, las filas seleccionadas ya no pueden repetirse
np.random.seed(1) 
for _ in range(int((len(df1)*60)/100)):  
    samples = np.random.choice(a=int((len(df)*60)/100), size=int((len(df)*60)/100), replace=False) 
bootstrappedDF1 = df1.loc[samples, :] 
```

```python
# Se grafica nuevamente la matriz de correlación
correlation= bootstrappedDF1.corr()
mask = np.triu(np.ones_like(correlation, dtype=bool))
f, ax = plt.subplots(figsize=(30, 30))
cmap = sns.diverging_palette(240, 10, n=9)
sns.heatmap(correlation, annot=True, mask=mask, cmap=cmap, vmax=.3, center=0,
             linewidths=.5, cbar_kws={"shrink": .5})
plt.yticks(fontsize=21)
plt.xticks(fontsize=21)
```

Nuevamente vemos que los datos no tienen una mayor correlación, pues en ambas situaciones las relaciones siguen siendo bastante débiles, y eso implica que simplemente no hay mayor relación que este tipo de gráfico pueda demostrar.
